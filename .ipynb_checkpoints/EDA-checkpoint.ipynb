```python
analysis_col = ['Store', 'Open', 'StoreType', 'DayOfWeek', 'Assortment', 'SchoolHoliday']
```

```python
X_train[analysis_col].head()
```

```python
print(test_data['Open'].isnull().sum())
print(test_data['Open'].value_counts())
```

```python
test_data['Open'] = test_data['Open'].fillna(1).astype(np.int64)
test_data['Store'] = test_data['Store'].astype('object')
test_data['DayOfWeek'] = test_data['DayOfWeek'].astype('object')
test_data[analysis_col].info()
test_data[analysis_col].head()
```

```python
embed_model = c2e.EmbeddingGenerator(epochs=25)
embed_model.fit(X_train[analysis_col], y_train)
```

```python
train_prediction = embed_model.predict(X_train[analysis_col])
print(len(train_prediction))
```

```python
test_prediction = embed_model.predict(test_data[analysis_col])
```

```python
train_transformed = embed_model.transform(X_train[analysis_col])
train_transformed.head()
```

```python
test_transformed = embed_model.transform(test_data[analysis_col])
test_transformed.head()
```

```python
def get_embeddings_plot(colname, transformed_data):
    col = [colname, f'{colname}_0', f'{colname}_1']
    x = transformed_data[col].drop_duplicates().sort_values(col).reset_index(drop=True)
    return x

plotdf = get_embeddings_plot('StoreType', train_transformed)
plotdf
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

var = 'Assortment'
b = pd.concat([X_train, y_train],axis=1)
c = pd.DataFrame(b.groupby(var)['Sales'].mean()).reset_index().rename({'Sales':'mean_sales'},axis=1)

fig, ax = plt.subplots(1,2, figsize=(12,5))
plotdf = get_embeddings_plot(var, train_transformed)
sns.scatterplot(data=plotdf, x=f'{var}_0', y=f'{var}_1', hue=var, palette='deep', s=50, ax=ax[0])
sns.barplot(data=c, x=var, y='mean_sales', ax=ax[1])
```

```python
import tensorflow as tf

# Number of categories and embedding dimension
n_categories = 4
embedding_dim = 3

# Define the embedding layer
embedding_layer = tf.keras.layers.Embedding(input_dim=n_categories, output_dim=embedding_dim)

# Example input indices
input_indices1 = tf.constant([0, 2, 1])
input_indices2 = tf.constant([1, 2, 0])

# Get the embedding vectors
embedding_vectors1 = embedding_layer(input_indices1)
embedding_vectors2 = embedding_layer(input_indices2)

print("Input indices:", input_indices1.numpy())
print("Embedding vectors:\n", embedding_vectors1.numpy())
print('\n')
print("Input indices:", input_indices2.numpy())
print("Embedding vectors:\n", embedding_vectors2.numpy())
```
